```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python

dataframe = pd.read_csv('Zomato data .csv')

```

```python
dataframe
```

```python

#REMOVE DENOMINATOR FROM RATING
def habdle_rating(value):
    value = str(value).split('/')
    return float(value[0])

dataframe['rate'] = dataframe['rate'].apply(habdle_rating)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of restaurant')
```

```python
grouped_data =dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel('Type of restaurant')
plt.xlabel('Votes',c="red", size=20)
```

```python
plt.hist(dataframe['rate'],bins=5)


#the majority of couples prefer resaurants with an approximate rating of 4.2 cost of 300 rupees
```

```python
#wheateronline orders received higher ratings offline orders.
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

```python
#conclusion:  offline order received less rating than online
```

```python
#Create the pivot table
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

# Create the heatmap
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')

# Set the title and labels
plt.title("Heatmap")
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')

# Show the plot
plt.show()
```
